```python
import json
import requests

# 大模型服务的地址
BASE_URL = "http://192.168.12.10:8000/v1/chat/completions"

# 请求参数
payload = {
  "model": "Qwen2.5-32B-Instruct-AWQ",
  "messages": [
    {
      "role": "user",
      "content": "请写一篇不少于10000字的文章，论述法学专业的就业前景。"
    }
  ],
  "stream": True  # 启用流式输出
}

# 发送请求并处理流式响应
try:
    with requests.post(BASE_URL, json=payload, stream=True, timeout=10) as response:
        if response.status_code == 200:
            print("请求成功，开始接收流式数据...")
            buffer = ""  # 用于存储不完整的数据块
            for line in response.iter_lines():
                if line:  # 过滤掉空行
                    line = line.decode("utf-8")  # 将字节数据解码为字符串
                    if line.startswith("data: "):  # 确保是有效的数据行
                        try:
                            chunk = json.loads(line[6:])  # 去掉 "data: " 前缀并解析 JSON
                            if "choices" in chunk and chunk["choices"]:
                                delta = chunk["choices"][0].get("delta", {})
                                content = delta.get("content", "")
                                print(content, end="", flush=True)  # 逐步打印内容
                        except json.JSONDecodeError:
                            print("跳过无效数据")
                            continue  # 跳过无效数据
        else:
            print(f"请求失败，状态码：{response.status_code}")
            print(response.text)
except requests.exceptions.RequestException as e:
    print(f"请求异常: {e}")
```
